```r
# Select customer records
sqldf("
SELECT *
FROM customers
LIMIT 5
")
```

```r
# Find customers from London
# The column names for 'customers' are: "customer_id", "age", "home_zone", "customer_type", "signup_date", "loyalty_score", "app_engagement_score", "preferred_channel", "account_status"
# It appears 'home_zone' is the appropriate column for location.
sqldf("
SELECT *
FROM customers
WHERE home_zone = 'London'
")
```

```r
# Count customers by home_zone
# The previous error indicated 'no such column: city'.
# Based on earlier analysis, 'home_zone' is the correct column name for location.
sqldf("
SELECT home_zone,
       COUNT(*) AS total_customers
FROM customers
GROUP BY home_zone
")
```

```r
# Join customers and orders
# The previous check confirmed that there is no 'order_status' column in the orders data frame.
# The available columns are: "order_id", "customer_id", "service_type", "order_created_at", "promised_window_hours", "pickup_zone", "dropoff_zone", "priority_level", "order_value", "booking_channel", "special_handling_flag"
# We will use 'service_type' as the appropriate column for order status.
sqldf("
SELECT customers.customer_id,
       orders.order_id,
       orders.service_type
FROM customers
JOIN orders
ON customers.customer_id = orders.customer_id
LIMIT 5
")
```

```r
# Count complaints by issue type
# The previous error indicated 'no such column: issue_type'.
# Based on the column names, 'complaint_type' is the appropriate column to use.
sqldf("
SELECT complaint_type,
       COUNT(*) AS total_complaints
FROM complaints
GROUP BY complaint_type
")
```

```r
# Retrieve only necessary columns
# The error 'no such column: city' means this column does not exist in the customers data frame.
# As previously identified, 'home_zone' is the correct column for location.
sqldf("
SELECT customer_id,
       home_zone
FROM customers
WHERE home_zone = 'Manchester'
")
```

```r
#Driver performance analysis
sqldf(
"SELECT driver_id,
       AVG(customer_rating_post_delivery) AS average_rating,
       COUNT(delivery_id) AS total_deliveries
FROM deliveries
GROUP BY driver_id
ORDER BY average_rating ASC
")
```

```r
# Route and operational zone analysis
sqldf(
"SELECT hub_id,
       COUNT(*) AS delayed_deliveries_count
FROM deliveries
WHERE delivery_status = 'Delayed'
GROUP BY hub_id
ORDER BY delayed_deliveries_count DESC
")
```
