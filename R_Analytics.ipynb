```r
 #install packages
 install.packages("ggplot2")
install.packages("dplyr")

library(ggplot2)
library(dplyr)
```

```r
#load dataset
customers <- read.csv("customers.csv")
orders <- read.csv("orders.csv")
deliveries <- read.csv("deliveries.csv")
complaints <- read.csv("complaints.csv")
drivers <- read.csv("drivers.csv")
```

```r
#Preview Data
head(customers)
```

```r
#summary statistic
summary(customers)
```

```r
#average loyalty score
mean(customers$loyalty_score)
```

```r
#failed delevaries count
table(deliveries$delivery_status)
```

```r
#complaint severity analysis
table(complaints$severity)
```

```r
#delivery status
ggplot(deliveries, aes(x = delivery_status)) +
  geom_bar() +
  ggtitle("Delivery Status Analysis")
```

```r
#complaint severity
severity_data <- table(complaints$severity)

pie(severity_data,
    main = "Complaint Severity Distribution")
```

```r
#driver ratings
hist(drivers$driver_rating,
     main = "Driver Ratings Distribution",
     xlab = "Ratings")
```

```r
table(orders$pickup_zone)
```

```r
# Print column names to check for typos or existence of 'total_orders'
print(names(customers))

loyalty_score_converted <- as.numeric(as.character(customers$loyalty_score))
total_orders_converted <- as.numeric(as.character(customers$total_orders))

# Print lengths of converted vectors to confirm which one is empty
print(paste("Length of loyalty_score_converted:", length(loyalty_score_converted)))
print(paste("Length of total_orders_converted:", length(total_orders_converted)))

customer_correlation_data <- data.frame(
    loyalty_score = loyalty_score_converted,
    total_orders = total_orders_converted
)

clean_data <- na.omit(customer_correlation_data)

cor(clean_data$loyalty_score,
    clean_data$total_orders)
```
