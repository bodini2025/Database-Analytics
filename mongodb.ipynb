```python
from pymongo import MongoClient

uri = "mongodb+srv://northstarAdmin:NorthStar123@northstarcluster.lrjpion.mongodb.net/?retryWrites=true&w=majority&appName=NorthStarCluster"

client = MongoClient(uri)

db = client["NorthStarDB"]

print("Connected Successfully")
```

```python
# Insert new customer document

new_customer = {
    "customer_id": 999,
    "name": "John Taylor",
    "city": "London",
    "loyalty_score": 88
}

db.customers.insert_one(new_customer)

print("Customer Inserted Successfully")
```

```python
# Read customer records

for x in db.customers.find().limit(5):
    print(x)
```

```python
# Update customer loyalty score

db.customers.update_one(
    {"customer_id": 999},
    {"$set": {"loyalty_score": 95}}
)

print("Customer Updated Successfully")
```

```python
# Delete customer document

db.customers.delete_one({"customer_id": 999})

print("Customer Deleted Successfully")
```

```python
# Aggregation analysis

pipeline = [
    {
        "$group": {
            "_id": "$city",
            "total_customers": {"$sum": 1}
        }
    }
]

result = db.customers.aggregate(pipeline)

for x in result:
    print(x)
```

```python
# Create single field index

db.customers.create_index("customer_id")

print("Single Field Index Created")
```

```python
# Create compound index

db.deliveries.create_index(
    [("city_zone", 1), ("delivery_status", 1)]
)

print("Compound Index Created")
```
